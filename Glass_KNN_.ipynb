```python
# Standardization using scaler min = 0, max = 1

def norm_fun(i):
  x = (i - i.min())/(i.max() - i.min())

  return x
```

```python
data_norm = norm_fun(data)
data = pd.DataFrame(data_norm)
data.describe()
```

```python
# Checking the Correalation

data.corr()
```

```python
# HeatMap

fig = plt.figure(figsize = (12,8))
sns.heatmap(data.corr(), cmap='viridis', annot = True) 
```

```python
data['Type'] = (data['Type'] > 0).astype(int)
```

```python
# Splitting the data

X = data.drop(columns = ['Type'])
Y = data.Type
```

```python
# Using the train test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
# Building The Model

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors = 5)
model.fit(x_train, y_train)

# Checking the model score

model.score(X, Y) # 82% is pretty good score 
```

```python
y_pred = model.predict(x_test)
```

```python
# Binarization - Discritization

y_test = (y_test > 0).astype(int)
y_pred = (y_pred > 0).astype(int)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
# Accuracy 
np.mean(y_pred==y_test)
```

```python
model.score(x_train,y_train)
```

```python
print("Accuracy", accuracy_score(y_test,y_pred)*100)
```

```python
#use cross validation score since this is a small size dataset 
#Get cross validation score of K-Nearest Neighbors

from sklearn.model_selection import cross_val_score
```

```python
X = (X > 0).astype(int)
Y = (Y > 0).astype(int)
```

```python
## K-fold cross validation
score = cross_val_score(model, X, Y, scoring='accuracy', cv = 15)
print(score)
print(score.mean())
```

```python
# Grid Search for Algorithm Tuning

n_neighbors = np.array(range(1,40))
param_grid = dict(n_neighbors=n_neighbors)
```

```python
from sklearn.model_selection import GridSearchCV

model1 = KNeighborsClassifier()
grid = GridSearchCV(estimator=model1, param_grid=param_grid)
grid.fit(X, Y)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
k_values = np.arange(1,25)
train_accuracy = []
test_accuracy = []

for i, k in enumerate(k_values):
    # k from 1 to 25(exclude)
    knn = KNeighborsClassifier(n_neighbors=k)
    # Fit with knn
    knn.fit(x_train,y_train)
    #train accuracy
    train_accuracy.append(knn.score(x_train,y_train))
    # test accuracy
    test_accuracy.append(knn.score(x_test,y_test))
# Plot
plt.figure(figsize=[13,8])
plt.plot(k_values, test_accuracy, label = 'Testing Accuracy')
plt.plot(k_values, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.title('-value VS Accuracy')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.xticks(k_values)
plt.savefig('graph.png')
plt.show()
print("Best accuracy is {} with K = {}".format(np.max(test_accuracy),1+test_accuracy.index(np.max(test_accuracy))))
```
